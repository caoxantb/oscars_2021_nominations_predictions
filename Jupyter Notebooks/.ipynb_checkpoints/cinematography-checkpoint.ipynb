```python
import pandas as pd
import numpy as np
import statistics
```

```python
import import_ipynb
from functions import df_to_dict, calc_percentage_2, calc_values, calc_bayes, visualize_prediction
```

```python
cine_oscars = pd.read_csv('../Best Cinematography/Extracted Data/oscar.csv', encoding='utf-8-sig')
cine_bafta = pd.read_csv('../Best Cinematography/Extracted Data/bafta.csv', encoding='latin-1')
cine_critics = pd.read_csv('../Best Cinematography/Extracted Data/critics.csv', encoding='latin-1')
cine_asc = pd.read_csv('../Best Cinematography/Extracted Data/asc.csv', encoding='utf-8-sig')
```

```python
cine_oscars_dict = df_to_dict(cine_oscars)
cine_bafta_dict = df_to_dict(cine_bafta)
cine_critics_dict = df_to_dict(cine_critics)
cine_asc_dict = df_to_dict(cine_asc)
```

```python
cine_awards = [cine_bafta_dict, 
                  cine_critics_dict, 
                  cine_asc_dict]

cine_union_all = []

for year in range(2010, 2020):
    union_all_each = cine_oscars_dict[year].union(cine_bafta_dict[year], 
                                                     cine_critics_dict[year], 
                                                     cine_asc_dict[year])
    cine_union_all.append(union_all_each)
    
cine_union_all
```

```python
cine_if, cine_mean = calc_values(cine_oscars_dict, cine_awards, cine_union_all)

# for year in range(2010, 2020):
#     print(len(cine_union_all[year-2010]))
    

cine_if
cine_mean
```

```python
cine_2020_bafta = pd.read_csv('../Best Cinematography/Applied Data/bafta.csv', encoding='utf-8-sig')
cine_2020_critics = pd.read_csv('../Best Cinematography/Applied Data/critics.csv', encoding='utf-8-sig')
cine_2020_asc = pd.read_csv('../Best Cinematography/Applied Data/asc.csv', encoding='utf-8-sig')
```

```python
cine_2020_bafta_dict = df_to_dict(cine_2020_bafta)[2020]
cine_2020_critics_dict = df_to_dict(cine_2020_critics)[2020]
cine_2020_asc_dict = df_to_dict(cine_2020_asc)[2020]
```

```python
cine_2020_union_all = cine_2020_bafta_dict.union(cine_2020_critics_dict, 
                                                       cine_2020_asc_dict)

cine_2020_union_all = list(cine_2020_union_all)

cine_2020_awards = [cine_2020_bafta_dict,
                       cine_2020_critics_dict,
                       cine_2020_asc_dict]


cine_2020_union_all
```

```python
cine_final, cine_prob_final = calc_bayes(cine_2020_awards, cine_2020_union_all, cine_mean)
cine_final
```
