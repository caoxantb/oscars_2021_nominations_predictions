```python
import pandas as pd
import statistics
from matplotlib import pyplot as plt
```

```python
import import_ipynb
from functions import df_to_dict, calc_percentage_2, calc_values, calc_bayes, visualize_prediction
```

```python
picture_oscars = pd.read_csv('../Best Picture/oscar.csv', encoding='utf-8-sig')
picture_bafta = pd.read_csv('../Best Picture/bafta.csv', encoding='utf-8-sig')
picture_critics = pd.read_csv('../Best Picture/critics.csv', encoding='utf-8-sig')
picture_pga = pd.read_csv('../Best Picture/pga.csv', encoding='utf-8-sig')
picture_sag = pd.read_csv('../Best Picture/sag.csv', encoding='utf-8-sig')
picture_globes = pd.read_csv('../Best Picture/globes.csv', encoding='utf-8-sig')
```

```python
picture_oscars_dict = df_to_dict(picture_oscars)
picture_bafta_dict = df_to_dict(picture_bafta)
picture_critics_dict = df_to_dict(picture_critics)
picture_pga_dict = df_to_dict(picture_pga)
picture_sag_dict = df_to_dict(picture_sag)
picture_globes_dict = df_to_dict(picture_globes)
```

```python
picture_awards = [picture_bafta_dict, 
                  picture_critics_dict, 
                  picture_pga_dict, 
                  picture_sag_dict, 
                  picture_globes_dict]

picture_union_all = []

for year in range(2010, 2020):
    union_all_each = picture_oscars_dict[year].union(picture_bafta_dict[year], 
                                                     picture_critics_dict[year], 
                                                     picture_pga_dict[year], 
                                                     picture_sag_dict[year],
                                                     picture_globes_dict[year])    
    picture_union_all.append(union_all_each)    
    
picture_if, picture_mean = calc_values(picture_oscars_dict, picture_awards, picture_union_all)

# for year in range(2010, 2020):
#     print(len(picture_union_all[year-2010]))
    

picture_if
picture_mean
```
