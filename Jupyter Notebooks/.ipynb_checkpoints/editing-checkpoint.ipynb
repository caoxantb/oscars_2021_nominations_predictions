```python
import pandas as pd
import numpy as np
import statistics
```

```python
import import_ipynb
from functions import df_to_dict, calc_percentage_2, calc_values, calc_bayes, visualize_prediction
```

```python
edit_oscars = pd.read_csv('../Best Editing/Extracted Data/oscar.csv', encoding='utf-8-sig')
edit_bafta = pd.read_csv('../Best Editing/Extracted Data/bafta.csv', encoding='latin-1')
edit_critics = pd.read_csv('../Best Editing/Extracted Data/critics.csv', encoding='latin-1')
edit_ace = pd.read_csv('../Best Editing/Extracted Data/ace.csv', encoding='utf-8-sig')
```

```python
edit_oscars_dict = df_to_dict(edit_oscars)
edit_bafta_dict = df_to_dict(edit_bafta)
edit_critics_dict = df_to_dict(edit_critics)
edit_ace_dict = df_to_dict(edit_ace)
```

```python
edit_awards = [edit_bafta_dict, 
                  edit_critics_dict, 
                  edit_ace_dict]

edit_union_all = []

for year in range(2010, 2020):
    union_all_each = edit_oscars_dict[year].union(edit_bafta_dict[year], 
                                                     edit_critics_dict[year], 
                                                     edit_ace_dict[year])
    edit_union_all.append(union_all_each)
    
edit_union_all
```

```python
edit_if, edit_mean = calc_values(edit_oscars_dict, edit_awards, edit_union_all)

# for year in range(2010, 2020):
#     print(len(edit_union_all[year-2010]))
    

edit_if
edit_mean
```

```python
edit_2020_union_all = edit_2020_bafta_dict.union(edit_2020_critics_dict, 
                                                       edit_2020_ace_dict)

edit_2020_union_all = list(edit_2020_union_all)

edit_2020_awards = [edit_2020_bafta_dict,
                       edit_2020_critics_dict,
                       edit_2020_ace_dict]


edit_2020_union_all
```
