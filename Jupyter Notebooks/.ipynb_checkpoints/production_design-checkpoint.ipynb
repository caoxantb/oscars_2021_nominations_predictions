```python
import pandas as pd
import numpy as np
import statistics
```

```python
import import_ipynb
from functions import df_to_dict, calc_percentage_2, calc_values, calc_bayes, visualize_prediction
```

```python
pde_oscars = pd.read_csv('../Best Production Design/Extracted Data/oscar.csv', encoding='utf-8-sig')
pde_bafta = pd.read_csv('../Best Production Design/Extracted Data/bafta.csv', encoding='latin-1')
pde_critics = pd.read_csv('../Best Production Design/Extracted Data/critics.csv', encoding='latin-1')
pde_adg = pd.read_csv('../Best Production Design/Extracted Data/adg.csv', encoding='latin-1')

pde_oscars_dict = df_to_dict(pde_oscars)
pde_bafta_dict = df_to_dict(pde_bafta)
pde_critics_dict = df_to_dict(pde_critics)
pde_adg_dict = df_to_dict(pde_adg)
```

```python
pde_awards = [pde_critics_dict, pde_bafta_dict, pde_adg_dict]

pde_union_all = []

for year in range(2010, 2020):
    union_all_each = pde_oscars_dict[year].union(pde_critics_dict[year], pde_bafta_dict[year], pde_adg_dict[year])
                                                
    pde_union_all.append(union_all_each)
    
pde_union_all
```

```python
pde_if, pde_mean = calc_values(pde_oscars_dict, pde_awards, pde_union_all)

# for year in range(2010, 2020):
#     print(len(pde_union_all[year-2010]))
    

pde_if
pde_mean
```
